```python
df.groupby('Product')[['Discount', 'Profit']].mean().sort_values('Discount', ascending=False)
```

```python
import matplotlib.pyplot as plt

product_summary = df.groupby('Product')[['Discount', 'Profit']].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 6))
colors = ['red' if p < 0 else 'steelblue' for p in product_summary['Profit']]
ax.barh(product_summary['Product'], product_summary['Profit'], color=colors)
ax.axvline(0, color='black', linewidth=0.8)
ax.set_xlabel('Average Profit per Deal ($)')
ax.set_title('Average Profit by Product (red = unprofitable)')
plt.tight_layout()
plt.show()
```
